```python
df.isna().sum()
```

**Agora as colunas que iremos usar como features não tem mais dados ausentes.**

**Separando os dados em features(x) e alvo(y).**

```python
x = df.drop(columns='birth_rate') # Features
y = df['birth_rate'] # Alvo

display(x)
display(y)

display(x.shape)
display(y.shape)
```

**Vamos para separação dos dados em treino e teste.**

```python
from sklearn.model_selection import train_test_split # Para separação em treino e teste
```

```python
# Vamos dividir em 30% teste e 70% treino
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)

print(x_treino.shape)
print(x_teste.shape)
print(y_treino.shape)
print(y_teste.shape)
```

**Vamos fazer a normalização dos dados.**

```python
from sklearn.preprocessing import StandardScaler # Para normalizar

# Criando objeto para normalizar
normalizador = StandardScaler()
```

```python
# Vamos "ensinar" para normalizar
normalizador.fit(x_treino)
```

```python
# Vai retornar um array
x_treino_normalizado = normalizador.transform(x_treino)
x_teste_normalizado = normalizador.transform(x_teste)

# Convertendo para DataFrame para visualizar melhor
df_x_treino_normalizado = pd.DataFrame(x_treino_normalizado, columns=x_treino.columns)
df_x_teste_normalizado = pd.DataFrame(x_teste_normalizado, columns=x_teste.columns)

display(df_x_treino_normalizado.head())
display(df_x_teste_normalizado.head())
```

**Vamos ver as escalas das variáveis com o boxplot.**

## **Modelos de Regressão**

```python
# Criando uma listas vazias para guardar os resultados
resultados_r2 = []
resultados_mae = []
resultados_mse = []
resultados_rmse = []
```

### **K-Nearest Neighbors (KNN)**

```python
# Importando o KNN regressor
from sklearn.neighbors import KNeighborsRegressor

n_vizinhos = KNeighborsRegressor(n_neighbors=3)
```

**Vamos treinar o modelo.**

```python
n_vizinhos.fit(x_treino_normalizado, y_treino)
```

**Previsão para os dados de teste.**

```python
y_pred_knn = n_vizinhos.predict(x_teste_normalizado)

y_pred_knn
```

**Vamos avaliar o modelo com uma métrica.**

```python
'''Métricas de avaliação do modelo (Performance) -> Erro Médio Absoluto (MAE)
Coeficiente de determinação (R²), Erro Quadrático Médio (MSE)'''
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Usando o (R²)
r2_knn = r2_score(y_teste, y_pred_knn)
resultados_r2.append(r2_knn)

# Usando o (MAE)
mae_knn = mean_absolute_error(y_teste, y_pred_knn)
resultados_mae.append(mae_knn)

# Usando o (MSE)
mse_knn = mean_squared_error(y_teste, y_pred_knn)
resultados_mse.append(mse_knn)

# Usando o (RMSE)
rmse_knn = np.sqrt(mse_knn)
resultados_rmse.append(rmse_knn)

print(f"R² (Score):          {r2_knn}")
print(f"MAE (Erro Médio):    {mae_knn}")
print(f"MSE (Erro  Quadrático Médio):    {mse_knn}")
print(f"RMSE (Erro Penal.):  {rmse_knn}")
```

### **Árvore de Decisão**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

arvore_decisao = DecisionTreeRegressor(random_state=42)
```

**Vamos treinar o modelo.**

```python
arvore_decisao.fit(x_treino_normalizado, y_treino)
```

```python
y_pred_arvore = arvore_decisao.predict(x_teste_normalizado)
y_pred_arvore
```

**Vamos avaliar usando as métricas.**

```python
r2_arvore = r2_score(y_teste, y_pred_arvore)
resultados_r2.append(r2_arvore)
mae_arvore = mean_absolute_error(y_teste, y_pred_arvore)
resultados_mae.append(mae_arvore)
mse_arvore = mean_squared_error(y_teste, y_pred_arvore)
resultados_mse.append(mse_arvore)
rmse_arvore = np.sqrt(mse_arvore)
resultados_rmse.append(rmse_arvore)

print(f"R² (Score):          {r2_arvore}")
print(f"MAE (Erro Médio):    {mae_arvore}")
print(f"MSE (Erro  Quadrático Médio):    {mse_arvore}")
print(f"RMSE (Erro Penal.):  {rmse_arvore}")
```

### **Regressão Linear**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Normalization
```

**Normalizar.**

```python
normalizador = Normalization(axis=-1)
normalizador.adapt(np.array(x_treino))
```

**Definindo o modelo de Keras, Regressão.**

```python
modelo_keras = keras.Sequential([normalizador,layers.Dense(units=1)])   # 1 neurônio, 'linear'
modelo_keras.summary()
```

**Compilando o modelo.**

```python
modelo_keras.compile(optimizer=tf.optimizers.Adam(learning_rate=0.1),
                    loss='mean_absolute_error')
```

**Treinar o modelo.**

```python
history = modelo_keras.fit(x_treino, y_treino, epochs=100, validation_split=0.2, verbose=0)
```

**Previsão do dados.**

```python
y_pred_keras_raw = modelo_keras.predict(x_teste)
y_pred_keras = y_pred_keras_raw.flatten()
```

**Avaliando com as métricas.**

```python
# Calculando as métricas
r2_keras = r2_score(y_teste, y_pred_keras) # R2
resultados_r2.append(r2_keras)
mae_keras = mean_absolute_error(y_teste, y_pred_keras) # MAE
resultados_mae.append(mae_keras)
mse_keras = mean_squared_error(y_teste, y_pred_keras) # MSE
resultados_mse.append(mse_keras)
rmse_keras = np.sqrt(mse_keras) # RMSE
resultados_rmse.append(rmse_keras)


print(f"R² (Score):          {r2_keras}")
print(f"MAE (Erro Médio):    {mae_keras}")
print(f"MSE (Erro  Quadrático Médio):    {mse_keras}")
print(f"RMSE (Erro Penal.):  {rmse_keras}")
```
